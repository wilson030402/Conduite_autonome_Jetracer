```python
import socket
import time

def check_internet_socket(host="8.8.8.8", port=53, timeout=3):
    try:
        socket.setdefaulttimeout(timeout)
        socket.socket(socket.AF_INET, socket.SOCK_STREAM).connect((host, port))
        print("Internet is up")
    except socket.error as ex:
        print("Internet is down")

if __name__ == "__main__":
    while True:
        check_internet_socket()
        time.sleep(0.5)
```

```python
import socket
import time

def quick_speed_test(server="google.com", port=80, buffer_size=4096):
    start_time = time.time()

    # Créer un socket et se connecter au serveur
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    sock.connect((server, port))

    # Envoi d'une requête HTTP HEAD simple
    request = f"HEAD / HTTP/1.1\r\nHost: {server}\r\n\r\n"
    sock.sendall(request.encode())

    # Réception de la réponse
    response = sock.recv(buffer_size)

    # Calcul du temps pris pour le test
    elapsed_time = time.time() - start_time

    # Fermer le socket
    sock.close()

    # Estimation approximative de la vitesse de téléchargement en Mbps
    data_size = len(response) * 8 / 1_000_000  # Convertir en Mbits
    speed = data_size / elapsed_time

    print(f"Test effectué en {elapsed_time:.4f} secondes")
    print(f"Vitesse estimée : {speed:.2f} Mbps")
    print("-" * 30)

if __name__ == "__main__":
    while True:
        quick_speed_test()
        time.sleep(1)  # Attendre 1 minute avant de refaire le test
```

```python
import socket
import time

def download_test(server="speed.hetzner.de", port=80, file_path="/100MB.bin", buffer_size=4096):
    # Créer un socket et se connecter au serveur
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    sock.connect((server, port))

    # Construire la requête HTTP GET pour télécharger le fichier
    request = f"GET {file_path} HTTP/1.1\r\nHost: {server}\r\nConnection: close\r\n\r\n"
    sock.sendall(request.encode())

    # Commencer le téléchargement et mesurer le temps
    start_time = time.time()
    total_data = 0

    while True:
        data = sock.recv(buffer_size)
        if not data:
            break
        total_data += len(data)

    elapsed_time = time.time() - start_time

    # Fermer le socket
    sock.close()

    # Calculer la vitesse de téléchargement en Mbps
    download_speed = (total_data * 8) / (elapsed_time * 1_000_000)  # Convertir en Mbps

    print(f"Test effectué en {elapsed_time:.4f} secondes")
    print(f"Vitesse de téléchargement : {download_speed:.2f} Mbps")
    print("-" * 30)

if __name__ == "__main__":
    while True:
        download_test()
        time.sleep(10)  # Attendre 1 minute avant de refaire le test

        
```

```python
import requests
import time

def download_speed_test(url="https://speed.hetzner.de/10MB.bin"):
    # Commencer la mesure du temps
    start_time = time.time()

    # Télécharger le fichier
    response = requests.get(url, stream=True)
    
    # Vérifier le statut de la réponse
    if response.status_code != 200:
        print("Erreur lors du téléchargement.")
        return
    
    # Calculer la taille des données téléchargées
    total_data = 0
    for chunk in response.iter_content(4096):
        total_data += len(chunk)
    
    # Mesurer le temps écoulé
    elapsed_time = time.time() - start_time

    # Vérifier si des données ont été reçues
    if total_data == 0:
        print("Aucune donnée reçue, débit indéterminé.")
        return
    
    # Calculer la vitesse de téléchargement en Mbps
    download_speed = (total_data * 8) / (elapsed_time * 1_000_000)  # Convertir en Mbps

    print(f"Test effectué en {elapsed_time:.4f} secondes")
    print(f"Données reçues : {total_data / 1_000_000:.2f} Mo")
    print(f"Vitesse de téléchargement : {download_speed:.2f} Mbps")
    print("-" * 30)

if __name__ == "__main__":
    while True:
        download_speed_test()
        #time.sleep(60)  # Attendre 1 minute avant de refaire le test
```

```python
import speedtest

def check_internet_speed(min_speed_mbps=20):
    try:
        st = speedtest.Speedtest()
        st.download()  # Effectue un test de débit descendant
        download_speed_mbps = st.results.download / 1_000_000  # Convertir en Mbps
        
        print(f"Download speed: {download_speed_mbps:.2f} Mbps")
        
        if download_speed_mbps >= min_speed_mbps:
            print("Internet speed is sufficient")
            return True
        else:
            print("Internet speed is too low")
            return False
    except Exception as ex:
        print(f"Failed to test internet speed: {ex}")
        return False
    
if __name__ == "__main__":
    while True:
        check_internet_speed()
```
